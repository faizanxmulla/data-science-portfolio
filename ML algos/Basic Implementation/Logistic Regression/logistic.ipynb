## Goal is to find a line such that 'P' for all training data is MAX
## this is k/a Max likelihood.

## Importing Libraries & getting Data 

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
from sklearn.datasets import make_blobs 
```

```python
X, y = make_blobs(n_samples = 1000 , centers =2 ,random_state=42)
```

```python
plt.scatter(X[:,0], X[:,1] , c=y)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

## Model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train , y_train)
```

```python
# two centers, so two coefficients
# intercept is theta not i.e sum of all intercepts
# x1,x2 etc. are the features.

model.coef_ , model.intercept_
```

```python
x1_sample = np.linspace(-6 ,7)
```

```python
colors = ["red","blue"]
y_colors = list(map(lambda c:colors[c],y))
```

```python
y_colors[:5] , y[:5]
```

```python
plt.scatter(X[:,0] , X[:,1] , c=y)

for i, (coef, intercept) in enumerate(zip(model.coef_, model.intercept_)):
    x2_sample = -coef[0]/coef[1] * x1_sample - intercept/coef[1]
    plt.plot(x1_sample, x2_sample, color=colors[i])
```

```python
model.score(X_test , y_test)
```
