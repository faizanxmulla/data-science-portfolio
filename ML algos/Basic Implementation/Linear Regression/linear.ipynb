### In LR , training our model means finding the value of m & c 

## Importing Data & getting data

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_regression
```

## Features

```python
bias = 200
X,y,coef = make_regression(n_features=1,bias = bias ,random_state=42,coef=True,noise=10)
```

```python
X.shape , y.shape
```

```python
# plotting data w/o noise
plt.scatter(X,y)
plt.show()
```

```python
# this is the m & c provided by sklearn 
coef,bias
```

```python
# Equation of line
y_gen = coef*X + bias
```

```python
plt.scatter(X,y)
plt.plot(X,y_gen)
```

## Model

```python
# above line is plotted using sklearn
# now plotting line using Linear-Regression

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X,y)
```

```python
model.coef_ , model.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,y_gen,label="Sklearn")
plt.plot(X  ,model.predict(X) , label="LR")
plt.legend()
plt.show()
```
