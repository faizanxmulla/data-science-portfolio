# Model Building

```python
data.columns
```

```python
X = data[['age', ' workclass', ' education', ' marital-status', ' occupation',
' relationship', ' race', ' hours-per-week', ' native-country',' income']].values 

y = data[' sex'].values
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## SVM

```python
from sklearn.svm import SVC 
from sklearn.metrics import accuracy_score

svc = SVC()
svc.fit(X_train, y_train)
```

```python
y_pred_svc = svc.predict(X_test)
print('Accuracy Score using SVM : ' ,accuracy_score(y_test ,y_pred_svc))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier 

rf = RandomForestClassifier()
rf.fit(X_train ,y_train)
```

```python
y_pred_rf = rf.predict(X_test)
print('Accuracy Score using Random Forest : ',accuracy_score(y_test ,y_pred_rf))
```

# Model Evaluation 

```python
# creating Dataframe to check which model was the best

models = pd.DataFrame({
    'Model': [ 'Random Forest', 'SVM'],
    'Score': [accuracy_score(y_test, y_pred_rf),
              accuracy_score(y_test, y_pred_svc),
              ]})

models.sort_values(by='Score', ascending=False)
```
