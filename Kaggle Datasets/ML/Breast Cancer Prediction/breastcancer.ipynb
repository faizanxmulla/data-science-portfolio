## PCA for Dimensional Reduction 

```python
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()

scaler.fit(data)
scaled_data = scaler.transform(data)
```

```python
from sklearn.decomposition import PCA 

pca = PCA(n_components=2)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
scaled_data.shape , x_pca.shape
```

```python
# creating Dataframe 
pca_data = pd.DataFrame(data=x_pca, columns=['First principle component', 'Second principle component'])

pca_data
```

## Visualizing PCA

```python
classes = ['Benign', 'Malignant']

plt.figure(figsize=(16,8))
scatter = plt.scatter(x_pca[:, 0], x_pca[:, 1], c=data['diagnosis'])

plt.xlabel("First principle component ", fontsize=12)
plt.ylabel("Second principle component ", fontsize=12)
plt.title("Principal Component Analysis of Breast Cancer", fontsize=15)

plt.legend(handles=scatter.legend_elements()[0], labels=classes)
```

# Model Building

```python
X = data.drop(columns=['diagnosis', 'id'])
y = data['diagnosis']

X.shape ,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler 
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression 
lr = LogisticRegression()
lr.fit(X_train ,y_train)
```

```python
y_predict_lr = lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score ,confusion_matrix  

print("Accuracy Score :", accuracy_score(y_test, y_predict_lr))
```

```python
sns.heatmap(confusion_matrix(y_test ,y_predict_lr) ,annot=True,fmt='.3g')
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier 
```

```python
# choosing best number of neighburs for our model
neighbors_with_accuracies = {}

for i in range(1,15):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train ,y_train)
    y_predict_knn =knn.predict(X_test)
    acc = accuracy_score(y_predict_knn ,y_test)
    neighbors_with_accuracies[i] = acc
```

```python
neighbors_with_accuracies
```

```python
max(neighbors_with_accuracies.values())
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
```

```python
y_predict_knn = knn.predict(X_test)
```

```python
print("Accuracy Score :", accuracy_score(y_test, y_predict_knn))
```

```python
sns.heatmap(confusion_matrix(y_test, y_predict_knn), annot=True, fmt='.3g')
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='gini')
dt.fit(X_train, y_train)
```

```python
y_predict_dt = dt.predict(X_test)
```

```python
print("Accuracy Score :", accuracy_score(y_test, y_predict_dt))
```

```python
sns.heatmap(confusion_matrix(y_test, y_predict_dt), annot=True, fmt='.3g')
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(criterion='entropy')
rf.fit(X_train ,y_train)
```

```python
y_predict_rf = rf.predict(X_test)
```

```python
print("Accuracy Score :", accuracy_score(y_test, y_predict_rf))
```

```python
sns.heatmap(confusion_matrix(y_test, y_predict_rf), annot=True, fmt='.3g')
```

## Support Vector Machine (SVM)

```python
from sklearn.svm import SVC 
svc = SVC()
svc.fit(X_train, y_train)
```

```python
y_predict_svc = svc.predict(X_test)
```

```python
print("Accuracy Score :", accuracy_score(y_test, y_predict_svc))
```

```python
sns.heatmap(confusion_matrix(y_test, y_predict_svc), annot=True, fmt='.3g')
```

## Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train, y_train)
```

```python
y_predict_nb = nb.predict(X_test)
```

# Model Evaluation

```python
# creating Dataframe to check which model was the best

models = pd.DataFrame({
    'Model': ['Logistic-Regression', 'KNN', 'Decision-Tree', 'Random-Forest', 'SVM','Naive-Bayes'],
    'Score': [accuracy_score(y_test, y_predict_lr),
            accuracy_score(y_test, y_predict_knn),
            accuracy_score(y_test, y_predict_dt),
            accuracy_score(y_test, y_predict_rf),
            accuracy_score(y_test, y_predict_svc),
            accuracy_score(y_test, y_predict_nb),
            ]})
    
models.sort_values(by='Score', ascending=False)
```
