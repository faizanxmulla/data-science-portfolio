```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
base_url = "https://kando.tech/investors/all"
parsed_data = []
```

```python
for page in range(0, 634):
    print(f'Scraping page {page} ....')

    response = requests.get(f"{base_url}?page={page}")

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')

        if "unexpected error" in soup.text.lower():
            print(f"Skipping page {page} due to unexpected error.")
            parsed_data.append({"Investor": None, "Country": None, "Profile": None, "Deals": None})
            continue

        table = soup.find('table')
        if not table:
            print(f"No table found on page {page}")
            parsed_data.append({"Investor": None, "Country": None, "Profile": None, "Deals": None})
            continue

        rows = table.find_all('tr')

        for r in rows:
            cols = r.find_all('td')
            if cols and len(cols) >= 4:
                row_data = {
                    "Investor": cols[1].text.strip(),
                    "Country": cols[2].text.strip(),
                    "Profile": cols[3].text.strip(),
                    "Deals": cols[4].text.strip(),
                }
                parsed_data.append(row_data)

    else:
        print(f'Failed to retreive website. Status code: {response.status_code}')
        parsed_data.append({"Investor": None, "Country": None, "Profile": None, "Deals": None})
```

```python
parsed_df = pd.DataFrame(parsed_data)

parsed_df.to_excel('investors_data.xlsx', index=False)
print('Excel file saved successfully!')
```
